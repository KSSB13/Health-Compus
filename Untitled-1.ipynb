```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('C:\\Users\\solva\\Desktop\\Health Compus\\Data set 2\\Student health Data.csv')
```

```python
# Convert 'Timestamp' to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d/%m/%Y %H:%M')

# Extracting year, month, day, hour, and minute from Timestamp for further analysis
df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month
df['Day'] = df['Timestamp'].dt.day
df['Hour'] = df['Timestamp'].dt.hour
df['Minute'] = df['Timestamp'].dt.minute

# Plotting distribution of entries over years
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Year', bins=20)
plt.title('Distribution of Entries Over Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()

# Plotting distribution of entries over months
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Month', bins=12)
plt.title('Distribution of Entries Over Months')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.show()

# Plotting distribution of entries over days
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Day', bins=31)
plt.title('Distribution of Entries Over Days')
plt.xlabel('Day')
plt.ylabel('Frequency')
plt.show()

# Plotting distribution of entries over hours
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Hour', bins=24)
plt.title('Distribution of Entries Over Hours')
plt.xlabel('Hour')
plt.ylabel('Frequency')
plt.show()

# Plotting distribution of entries over minutes
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Minute', bins=60)
plt.title('Distribution of Entries Over Minutes')
plt.xlabel('Minute')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate BMI (Body Mass Index)
df['BMI'] = df['Weight(in KGs)'] / ((df['Height(in CMs/ Inches)'] / 100) ** 2)

# Plotting BMI distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='BMI', bins=20, kde=True)
plt.title('BMI Distribution')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Count')
plt.show()

# Print BMI statistics
print("BMI Statistics:")
print(df['BMI'].describe())
```

```python
# Convert 'Timestamp' to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Extract month from Timestamp
df['Month'] = df['Timestamp'].dt.month

# Map month numbers to month names for better visualization
month_map = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
df['Month'] = df['Month'].map(month_map)

# Count occurrences of diseases for each month
disease_counts_by_month = df.groupby('Month')['Purpose of visiting hospital'].value_counts().unstack().fillna(0)

# Plotting disease occurrences by month
plt.figure(figsize=(12, 8))
sns.heatmap(disease_counts_by_month, cmap='viridis', annot=True, fmt='g')
plt.title('Disease Occurrences by Month')
plt.xlabel('Purpose of visiting hospital')
plt.ylabel('Month')
plt.show()
```
